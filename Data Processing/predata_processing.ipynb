```python

product = pd.read_csv("/content/drive/MyDrive/AI4Good/Final Project/product_data.csv")
product
```

```python
# Filtering the DataFrame for each store
save_on_foods = product[product['Store'] == 'Save On Foods']
superstore = product[product['Store'] == 'Superstore']  # Assuming 'Superstore' is mentioned in the data
```

```python
save_on_foods
```

```python
#save_on_foods
def split_package_sizing(row):
    if pd.isna(row['Package Sizing']):
        # If the package sizing is NaN, use the Price as price per unit and set unit to '1'
        return pd.Series([row['Price'], '1ea'])
    else:
        parts = row['Package Sizing'].split('/')
        if len(parts) == 2:
            # If the split is successful and we have two parts
            price_per_unit, unit = parts
        else:
            # If the split does not result in exactly two parts, fall back to defaults
            price_per_unit = row['Price']  # Use the Price as a fallback
            unit = '1'  # Assume the unit is 1
        return pd.Series([price_per_unit, unit])

# Applying the function to the save_on_foods DataFrame
save_on_foods[['Price per Unit', 'Unit']] = save_on_foods.apply(split_package_sizing, axis=1)

# Showing the updated DataFrame to verify results
print(save_on_foods.head())
```

```python
save_on_foods['Price per Unit'] = save_on_foods['Price per Unit'].astype(str).str.replace('$', '')
```

```python
save_on_foods
```

```python

def split_units(unit_value):
    # Ensure the unit_value is a string, this avoids the TypeError
    if isinstance(unit_value, str):
        quantity = ''.join(filter(str.isdigit, unit_value))
        unit = ''.join(filter(str.isalpha, unit_value))
    else:
        quantity = ''
        unit = ''
    return pd.Series([quantity, unit])

# Now applying the function directly on the 'Unit' column rather than passing the row

# Applying the function to split the 'Unit' column
save_on_foods[['Per Unit Quantity', 'Unit Type']] = save_on_foods['Unit'].apply(split_units)
save_on_foods
```

```python
save_on_foods.drop('Unit', axis=1, inplace=True)
save_on_foods.drop('Package Sizing', axis=1, inplace=True)
```

```python
save_on_foods.to_csv('/content/drive/MyDrive/AI4Good/Final Project/SaveOnFoods_cleaned.csv', index=False)
```

```python
no_frills = product[product['Store'] == 'No Frills']
no_frills
```

```python
# Try extracting with the regular expression
no_frills[['Price Per Kilo', 'Price per Unit']] = no_frills['Package Sizing'].str.extract(r'([\d\.]+/\w+)\s*(\$\d+\.?\d*/\w+)')
no_frills['Quantity'] = None
# For rows where data is missing, split by comma as a fallback
fallback_df = no_frills[no_frills['Price Per Kilo'].isnull() | no_frills['Price per Unit'].isnull()]
for index, row in fallback_df.iterrows():
    package_sizing = str(row['Package Sizing'])
    if ',' in package_sizing:
        quantity, price = row['Package Sizing'].split(',', 1)  # Split at the first comma
        no_frills.at[index, 'Quantity'] = quantity.strip()
        no_frills.at[index, 'Price per Unit'] = price.strip()

no_frills.drop('Package Sizing', axis=1, inplace=True)
no_frills.drop('Price Per Kilo', axis=1, inplace=True)
no_frills.drop('Quantity', axis=1, inplace=True)

# Check the DataFrame to ensure the new columns are populated correctly
no_frills
```
